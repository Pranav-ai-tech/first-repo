```python
import pandas as pd

# ==============================
# 1. Load CSV
# ==============================
df = pd.read_csv("student_performance.csv")

# ==============================
# 2. Convert Timestamp
# ==============================
df["Timestamp"] = pd.to_datetime(df["Timestamp"], errors="coerce")

print("Before cleaning:", df.shape)

# ==============================
# 3. Remove rows without Timestamp
# ==============================
df_clean = df.dropna(subset=["Timestamp"])

print("After removing NaT timestamps:", df_clean.shape)

# ==============================
# 4. Sort by latest Timestamp
# ==============================
df_sorted = df_clean.sort_values(by="Timestamp", ascending=False)

# ==============================
# 5. Get latest N records
# ==============================
N = 5   # üîÅ change this to 1, 3, 10, etc.
latest_records = df_sorted.head(N)

latest_records
```
